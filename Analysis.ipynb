```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('combined_PO8_sym.csv')
df.set_index('time', inplace=True)
df
```

```python
plt.figure(figsize=(16, 12))
for col in df.columns[:-1]:  # Exclude the 'Average' column
    plt.plot(df.index, df[col], alpha=0.5, label=col)  # Lower opacity

# Highlight the average column
plt.plot(df.index, df['Avg_PO8'], color='black', linewidth=2, label='Average', alpha=1)
plt.xlabel('Time')
plt.ylabel('Aplitude')
plt.title('PO8 for Symmetry Pattern')
plt.axhline(y=0, color='red', linestyle='-', linewidth=2)
plt.legend(loc='best');
```
